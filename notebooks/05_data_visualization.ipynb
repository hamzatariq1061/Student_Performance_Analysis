```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/StudentsPerformance_cleaned.csv")

df["average score"] = df[["math score", "reading score", "writing score"]].mean(axis=1)

df.head()
```

```python
plt.figure(figsize=(8, 5))

sns.histplot(df["average score"], kde=True)

plt.title("Distribution of Average Student Scores")
plt.xlabel("Average Score")
plt.ylabel("Frequency")
plt.tight_layout()
plt.savefig("../outputs/histogram_kde_average_score.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8, 5))

sns.boxplot(x="gender", y="average score", data=df)

plt.title("Average Score Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.tight_layout()
plt.savefig("../outputs/boxplot_gender_average_score.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8, 5))

sns.scatterplot(
    x="reading score",
    y="writing score",
    hue="gender",
    data=df
)

plt.title("Relationship Between Reading and Writing Scores")
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.tight_layout()
plt.savefig("../outputs/scatter_reading_writing.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8, 5))

test_prep_avg = df.groupby("test preparation course")["average score"].mean().reset_index()

sns.barplot(
    x="test preparation course",
    y="average score",
    data=test_prep_avg
)

plt.title("Average Score by Test Preparation Course")
plt.xlabel("Test Preparation Course")
plt.ylabel("Average Score")
plt.tight_layout()
plt.savefig("../outputs/bar_test_preparation_average_score.png", dpi=300)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.histplot(df["average score"], kde=True, ax=axes[0, 0])
axes[0, 0].set_title("Histogram with KDE")

sns.boxplot(x="gender", y="average score", data=df, ax=axes[0, 1])
axes[0, 1].set_title("Boxplot by Gender")

sns.scatterplot(x="reading score", y="writing score", hue="gender", data=df, ax=axes[1, 0])
axes[1, 0].set_title("Scatter Plot")

sns.barplot(x="test preparation course", y="average score", data=test_prep_avg, ax=axes[1, 1])
axes[1, 1].set_title("Bar Chart by Test Preparation")

plt.tight_layout()
plt.savefig("../outputs/combined_visualizations.png", dpi=300)
plt.show()
```

## Data Visualization Observations

1. The histogram with KDE shows the overall distribution of students' average scores.
2. The boxplot compares the score distribution of male and female students.
3. The scatter plot shows a strong positive relationship between reading and writing scores.
4. The bar chart shows that students who completed the test preparation course have a higher average score.
5. The combined 2x2 subplot summarizes all major visual findings in one figure.
