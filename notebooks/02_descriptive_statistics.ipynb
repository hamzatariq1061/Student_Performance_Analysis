```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/StudentsPerformance.csv")
df.head()
```

```python
df.shape
```

```python
numeric_cols = ["math score", "reading score", "writing score"]

stats = pd.DataFrame({
    "Mean": df[numeric_cols].mean(),
    "Median": df[numeric_cols].median(),
    "Mode": df[numeric_cols].mode().iloc[0],
    "Standard Deviation": df[numeric_cols].std(),
    "Minimum": df[numeric_cols].min(),
    "Maximum": df[numeric_cols].max()
})

stats
```

```python
mean_median_comparison = pd.DataFrame({
    "Mean": df[numeric_cols].mean(),
    "Median": df[numeric_cols].median(),
    "Difference": df[numeric_cols].mean() - df[numeric_cols].median()
})

mean_median_comparison
```

```python
df.groupby("gender")[numeric_cols].mean()
```

```python
df.groupby("test preparation course")[numeric_cols].mean()
```

## Observations

1. The average reading score is higher than the average math score.
2. The average writing score is higher than the average math score.
3. The mean and median values are close, which shows that the score distribution is not highly skewed.
4. Students who completed the test preparation course performed better on average than students who did not complete it.
5. GroupBy analysis by gender shows differences in average performance across math, reading, and writing scores.
